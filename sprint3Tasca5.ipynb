```python
# Localizar valores nulos
temp_df.isnull().sum()
```

```python
# Eliminar lineas donde hay un % bajo de valores nulos en algunas columnas.
temp_df = temp_df.dropna(axis=0, subset=['ArrTime','TailNum','ActualElapsedTime','CRSElapsedTime','AirTime','ArrDelay'])
temp_df.isnull().sum()
```

```python
# Hay muchos valores nulos en las colunas que restan (cerca de 35% de total) y las las columnas me interesan
# No voy a eliminar todas estas colunmas ni las líneas donde hay valores nulos porque son interesantes. 
# Entonces voy a sustituir los valores nulos por zero.
temp_df = temp_df.fillna(0)
temp_df.isnull().sum()
```
